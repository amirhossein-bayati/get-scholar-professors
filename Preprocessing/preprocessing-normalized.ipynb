# Import Libraries

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing

%matplotlib inline
```

# Load The Dataset

```python
df = pd.read_excel('../Data-gathering/Json/Final_Dataset.xlsx')
df = df.drop(['Index',
           'Date',
           'index',
           'Author name',
           'University',
           'country',
           'Google Scholar link of the university',
           'Google Scholar link of the author',
           'Qs uni world ranking',
           'Qs uni country ranking',
           'Date of the first publication (year)',
           'Date of the last publication (year)',
            ], axis=1)
cols = df.columns
main_df = df
```

# Implement Scalers and Normalizer on final dataset

## MinMax Scaler

```python
scaler = preprocessing.MinMaxScaler()
scaled = scaler.fit_transform(main_df)
min_maxed_df = pd.DataFrame(scaled, columns=[cols])
min_maxed_df
```

## Standard Scaler

```python
scaler = preprocessing.StandardScaler()
scaled = scaler.fit_transform(min_maxed_df)
df = pd.DataFrame(scaled, columns=[cols])
df
```

## Normalization

```python
df_normalized = preprocessing.normalize(min_maxed_df)
df = pd.DataFrame(df_normalized, columns=[cols])
df
```

# Statistical parameters

```python
statistical = []
for col in df.columns:
    statistical_par = {
            'Minimum': round(df[col].min(),2),
            'Maximum': round(df[col].max(),2),
            'Median':  round(df[col].median(),2),
            'Mean':    round(df[col].mean(),2),
            'STD':     round(df[col].std(),2),
            'Variance':round(df[col].var(),2)
        }
    statistical.append(statistical_par)
pd.DataFrame(
    statistical, 
    index = cols
)
```

# Histograms

```python
fig, ax = plt.subplots(nrows=5, ncols=4, figsize=(20, 20))
i = 0
j = 0
for col in cols:
    if j == 4:
        i += 1
        j = 0
    ax[i, j].set(title=col)
    ax[i, j].hist(df[col], 30)
    j += 1
```

# Density Plot

```python
for col in cols:
    df[col].plot(kind='density', subplots=True, title=col)
    plt.show()
```

# Box Plots

```python
for col in cols:
    sns.boxenplot(x=main_df[col])
    plt.show()
```

# Correlation Heatmap

```python
plt.figure(figsize=(15,15))
corrMatrix = df.corr()
sns.heatmap(corrMatrix, annot=True, fmt="5.2f")
plt.show()
```
